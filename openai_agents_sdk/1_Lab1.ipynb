```python
from dotenv import load_dotenv
from agents import Agent, trace, Runner
```

```python
load_dotenv(override=True)
```

```python
math_tutor = Agent(name="Math_Tutor", instructions="You are an expert mathematician to solve Maths problem", model="gpt-4o-mini")
```

```python
math_question = f"""
Tickets numbered 1 to 20 are mixed up and then a ticket is drawn at random. 
What is the probability that the ticket drawn has a number which is a multiple of 3 or 5?
"""
response = await Runner.run(math_tutor,input=math_question)
print(response.final_output)
```

```python
jokester = Agent(name="Jokester", instructions="You are a funny joke teller", model="gpt-4o-mini")
```

```python
with trace("telling a joke:"):
    response = await Runner.run(jokester,"Tell me a dirty joke")
    print(response.final_output)
```
