```python
# read the pdf file
reader = PdfReader("me\SumanPrakash_LinkedIn.pdf")
linkedIn_profile = ""
for page in reader.pages:
    text = page.extract_text()
    if text:
        linkedIn_profile += text
```

```python
# Creating a prompt for answering any queries about me using linkedin profile and summary as source
name = "Suman Prakash"
system_prompt = f"""You are acting as {name}. You are answering questions on {name}'s website, 
particularly questions related to {name}'s career, background, skills and experience. 
Your responsibility is to represent {name} for interactions on the website as faithfully as possible. 
You are given a summary of {name}'s background and LinkedIn profile which you can use to answer questions. 
Be professional and engaging, as if talking to a potential client or future employer who came across the website. 
If you don't know the answer, say so.

With this context, please chat with the user, always staying in character as {name}.

PFB the summary and Linkedin Profile as resource that you need to refer for answering questions or 
communicating with the user on other side.

## Summary:\n{summary}\n\n
## LinkedIn Profile:\n{linkedIn_profile}
"""
```

```python
from IPython.display import display, Markdown
display(Markdown(system_prompt))
```

```python
# Create a "chat" function that should accept chat message and history, and return the response from AI for the message 
# keeping system prompt in context

def chat(message, history):
    request = [{"role":"system", "content":system_prompt}] + history + [{"role":"user","content":message}]
    response = openai_client.chat.completions.create(model="gpt-4o-mini", messages=request)
    return response.choices[0].message.content
```

```python
#launch a messaging UI through gradio using the chat function
gr.ChatInterface(fn=chat).launch()
```

```python
# Create an Evaluation class using BaseModel from Pydantic to retrieve response from Evaluator LLM in this particular format
from pydantic import BaseModel
class Evaluation(BaseModel):
    isacceptable:bool
    feedback:str
```

```python
# Create a system prompt for Evaluator
evaluator_system_prompt = f""" You are an evaluator that decides whether the response to a question is acceptable.\n
You are provided with a conversation between a user and an agent. And, your task is to decide whether the agent's latest response is acceptable quality.\n
The agent is playing the role of {name} and is representing {name} on their website. \n
The agent has been instructed to be professional and engaging, as if talking to potential client or future employer who came across the website.\n
The agent has been provided with the context on {name} via their summary and LinkedIN profile that you can find below:\n
##Summary:\n {summary}\n
##LinkedinProfile:\n {linkedIn_profile}\n

With this context, please evaluate the latest response of the agent in the conversation and reply with whether it is acceptable and your feedback.

"""
```

```python
display(Markdown(evaluator_system_prompt))
```

```python
# Create a method to design the Evaluator user prompt that receives the latest reply from agent, message from user,
#  and the coversation history as parameter

def evaluator_user_prompt(agent_reply, user_message, chat_history):
    evaluator_user_prompt= f""" 
    Here's the conversation between the user and the agent:\n\n{chat_history}\n\n
    Here's the latest message from the user: \n\n{user_message}\n\n
    Here's the latest reply from the agent: \n\n{agent_reply}\n\n
    Now, Please evaluate the latest reply from the agent and respond whether it is acceptable with your feedback.
    """
    return evaluator_user_prompt
```

```python
# Create a client to call Gemini API
import os
gemini_api_key = os.getenv("GOOGLE_API_KEY")
gemini = OpenAI(
    api_key=gemini_api_key,
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)
```

```python
# Create an evaluate function in which we are going to call gemini-2.5-flash model to evaluate the agent's reply 
# and capture response in Evaluation class format
def evaluate(reply, message, history) -> Evaluation: # We want the return from this function as Evaluation
    #Note we have called the evaluator_user_prompt function to initialize messages variable and passing the same parameters from Evaluate function
    messages=[{"role":"system", "content":evaluator_system_prompt}] + [{"role":"user", "content":evaluator_user_prompt(reply, message, history)}]
    #Note the change in calling the Gemini API, we have used .parse and one of the parameter we added is response_format
    #  to ensure the response comes in specified format
    response = gemini.beta.chat.completions.parse(
        model="gemini-2.5-flash",
        messages=messages,
        response_format=Evaluation
    )
    # Note the below line, we called .parsed to render the response in Evaluation class format
    return response.choices[0].message.parsed
```

```python
#Create a Rerun function to rerun the response from GPT-4o-mini in case the feedback is negative
def rerun(agent_reply, user_message, chat_history, feedback):
    updated_system_prompt = f""" {system_prompt}\n\n 
    ## Previous answer rejected: You just tried to answer the question. However, Quality control rejected your answer on user message.\n\n
    ## Message from User : {user_message}\n\n
    ## Your attempted answer: {agent_reply}\n\n
    ## QA feedback for rejection: {feedback}\n\n
    Please attempt to answer the question properly considering the feedback from QA."""
    message = [{"role":"system", "content":updated_system_prompt}] + chat_history + [{"role":"user", "content":user_message}]
    response = openai_client.chat.completions.create(
        model="gpt-4o-mini",
        messages=message
    )
    return response.choices[0].message.content
```

```python
# Creating a single workflow for answering the question, evaluating it, if evaluation fails, re-run the response
def chat(message, history):
    if "patent" in message:
        system = f"{system_prompt}\n\n Everything in your reply needs to be in Pig Latin\
            It is mandatory that you reply only and entirely in Pig Latin."
    else:
        system = system_prompt
    message = [{"role":"system", "content":system}] + history + [{"role":"user", "content":message}]
    response = openai_client.chat.completions.create(
        model="gpt-4o-mini",
        messages=message
    )
    reply = response.choices[0].message.content
    # Evaluate the reply
    evaluation = evaluate(reply,message,history)
    # if evaluation fail rerun the response
    if evaluation.isacceptable:
        print("Passed Evaluation- Returning Reply")
        print(history, message)
    else:
        print(f"Failed Evaluation- Retrying\n\n Feedback: {evaluation.feedback}")
        reply = rerun(reply, message, history, evaluation.feedback)
    return reply
        

```

```python
gr.ChatInterface(chat).launch()
```
